 Use beautiful Soup to scrape a webpage of your choice and extract clean information from it.

```python
# BEautiful Soup

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests
import csv


#Working with data from the official website of the National Gallery of Art in the United States
# I'm going to pull limited data from this site ( Artists names starting letter 'J')

# Collect first page of artistsâ€™ list
page = requests.get('https://web.archive.org/web/20121007172955/https://www.nga.gov/collection/anJ1.htm')

# Create a BeautifulSoup object
b_soup = BeautifulSoup(page.text, 'html.parser')

# Removing the unwanted information, the links which does not navigate to artists information
#decomposing the "AlphaNav" class

last_links = b_soup.find(class_='AlphaNav')
last_links.decompose()

# Create a .csv file to write to, add headers row
f = csv.writer(open('J-artist-names_BeautifulSoup.csv', 'w'))
f.writerow(['Name', 'Link'])

# Pull all text from the BodyText div
artist_name_list = b_soup.find(class_='BodyText')

# Pull text from all instances of <a>- hyperlink tag tag within BodyText div
artist_name_list_items = artist_name_list.find_all('a')

# Create for loop to print out only  artists' names from <a> links
for artist_name in artist_name_list_items:
   # print(artist_name.prettify())
    names = artist_name.contents[0]  # get the artists names
    links = 'https://web.archive.org' + artist_name.get('href')  # get the hyperlink of the artist names
    print (names)
    print (links)
    f.writerow([names, links]) # Writing the data into .csv file
```
